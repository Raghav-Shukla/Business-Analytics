```python
import numpy as np
import matplotlib.pyplot as p
from mpl_toolkits import mplot3d
fig = p.figure()
ax = p.axes(projection="3d")
x=[0,1,2,3,4,5,6]
y=[0,1,4,9,16,25,36]
z=[0,1,8,27,64,125,216]
p.xlabel("X axis")
p.ylabel("Y axis")
p.title("3D Plot")
 
ax.plot3D(x, y, z, 'red')
ax.scatter3D(x, y, z, c=z, cmap='plasma');
p.show()
```

```python
from matplotlib import pyplot as p
import numpy as np
softdrinks = ['Pepsi', 'Coca-Cola', 'Mirinda',
        'Sprite', 'Appy Fizz', 'Frooti'] 
data = [48, 23, 89, 21, 62, 47]
fig = p.figure(figsize =(10, 7))
p.pie(data, labels = softdrinks)
p.title("Pie-chart")
p.show()
```

```python
import matplotlib.pyplot as p
import numpy as np
x = np.linspace(-2, 2, 1000)
y = (x ** 3)-(3*x)


fig = p.figure(figsize = (10, 5))
p.xlabel("X axis")
p.ylabel("Y axis")
p.title("My favourite function from JEE times")
p.plot(x, y)
p.show()
```
