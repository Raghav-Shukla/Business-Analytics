```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import preprocessing, svm
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
df=pd.read_excel('Regression Data.xlsx')
```

```python
df
```

```python
#This one is for linear regression using volume

df_binary1 = df[['Volume', 'CO2']]

X = np.array(df_binary1['Volume']).reshape(-1, 1)     # Separating the data into independent and dependent variables
y = np.array(df_binary1['CO2']).reshape(-1, 1)        # Converting each dataframe into a numpy array

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size =0.2)
 
regr = LinearRegression()
 
regr.fit(X_train, y_train)
print(regr.score(X_test, y_test))
```

```python
y_pred = regr.predict(X_test)
plt.scatter(X_test, y_test, color ='b')
plt.plot(X_test, y_pred, color ='k')
 
plt.show()
```

```python
#This one is for linear regression using weight

df_binary2 = df[['Weight', 'CO2']]

X1 = np.array(df_binary2['Weight']).reshape(-1, 1)     # Separating the data into independent and dependent variables
y1 = np.array(df_binary2['CO2']).reshape(-1, 1)        # Converting each dataframe into a numpy array

X_train1, X_test1, y_train1, y_test1 = train_test_split(X1, y1, test_size =0.2)
 
regr = LinearRegression()
 
regr.fit(X_train1, y_train1)
print(regr.score(X_test1, y_test1))
```

```python
y_pred1 = regr.predict(X_test1)
plt.scatter(X_test1, y_test1, color ='b')
plt.plot(X_test1, y_pred1, color ='k')
 
plt.show()
```
