```python
df1=df.dropna(subset=['Weight'])
df1
```

```python
import numpy as np
df2=df1.fillna({'DOB':"unknown"})
df2
```

```python
g1=df2.groupby('Pos')
g1
```

```python
df3=df2
```

```python
#Function to get median height 
def convert_height(cell):
  x=cell.split("\'")
  th=(float(x[0])*12)+float(x[1])
  return th

df3['Height']=df3['Height'].apply(convert_height)
df3

#I converted each and every element of the column's string variable into an effective height
```

```python
g2=df3.groupby('Team')
g2
```

```python
x=g2.get_group('Men').Height.median()
print(str(int(x/12))+'\''+str(int(x%12)))
```

```python
def convert_weight(cell):
  return float(cell)  
df3['Weight']=df3['Weight'].apply(convert_weight)  
```
