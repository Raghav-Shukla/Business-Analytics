```python
# Now we do the last question
# We imprt all the necessary libraries
import sklearn
import numpy as np
from sklearn import linear_model
from sklearn import metrics
from sklearn.model_selection import train_test_split
```

```python
X = np.array(df['tempo']).reshape(-1,1)
y = np.array(df['label']).reshape(-1,1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 1)  #To split the data into test data and train data
digreg = linear_model.LogisticRegression()
digreg.fit(X_train, y_train)
```

```python
# Accuracy of score
y_pred = digreg.predict(X_test)
print("Accuracy of Logistic Regression model is:",
metrics.accuracy_score(y_test, y_pred)*100)
```
