```python
#Form k clusters by assigning each instance to its nearest centroid.
#Recompute the centroid of each cluster.
#In this section, we perform k-means clustering on a toy example of movie ratings dataset. 
import pandas as pd

ratings = [['john',5,5,2,1],['mary',4,5,3,2],['bob',4,4,4,3],['lisa',2,2,4,5],['lee',1,2,3,4],['harry',2,1,5,5]]
titles = ['user','Jaws','Star Wars','Exorcist','Omen']
movies = pd.DataFrame(ratings,columns=titles)
movies
```

```python
# apply k-means clustering (with k=2) on the movie ratings data.
# You must must remove the "user" column first before applying the clustering algorithm. The cluster assignment for each user is displayed as a dataframe object.
```

```python
movies_n=movies.drop('user',axis=1)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns 
import numpy as np
from sklearn.cluster import KMeans
from mpl_toolkits.mplot3d import Axes3D
```

```python
kmeans = KMeans(n_clusters=2, random_state=0)
kmeans.fit(movies_n)
kmeans.cluster_centers_
```

```python
#This is a plot showing the 4d points. The first 3 dimensions are shown in the 3d plot. The last dimension is explained by means of the colorbar
#The points are very less and small. If i plot the cluster centers right now itself, you may not be able to differentiate. So i have plotted them in green in the next cell.
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x=np.array(movies_n['Jaws']).reshape(-1,1)
y=np.array(movies_n['Star Wars']).reshape(-1,1)
z=np.array(movies_n['Exorcist']).reshape(-1,1)
c=np.array(movies_n['Omen']).reshape(-1,1)
img = ax.scatter(x, y, z, c=c, cmap=plt.hot())

fig.colorbar(img)
plt
plt.show()
```
